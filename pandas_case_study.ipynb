#Q.8 Handle missing values in 'Revenue (Millions)' and 'Metascore


```python
median_value1 = df['Revenue (Millions)'].median()
median_value1
```

```python
median_value2 = df['Metascore'].median()
median_value2
```

```python
df['Revenue (Millions)'].fillna(median_value1, inplace=True)
df['Metascore'].fillna(median_value2, inplace=True)
df.info()
```

# Q.17 Create a column that categorizes movies by rating:
a) Rating < 6: Poor



```python
df.loc[df['Rating'] < 6, 'Category'] = 'Poor'
print(df[df['Rating'] < 6][['Title', 'Rating', 'Category']])
```

b) 6 â‰¤ Rating < 7.5: Average

```python
average_movies = df[(df['Rating'] >= 6) & (df['Rating'] < 7.5)][['Title', 'Rating']]
average_movies['Category'] = 'Average'

print(average_movies)
```

c) Rating â‰¥ 7.5: Excellent

```python
excellent_movies = df[df['Rating'] >= 7.5][['Title', 'Rating']]
excellent_movies['Category'] = 'Excellent'

print(excellent_movies)
```

# Q.18 Create a custom function to score movies based on:
i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
Apply a function to dataset to assign the score

```python
def score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

df['High_Score'] = df.apply(score, axis=1)
df[['Title','Rating','Revenue (Millions)','High_Score']]
```
